```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('image_search_results.json') as json_file:
    image_resutls = json.load(json_file)

keys = list(map(int,image_resutls.keys()))
keys.sort()
test = [image_resutls[str(key)][0] for key in keys]
train = [image_resutls[str(key)][1] for key in keys]

labels =  [str(int(np.sqrt(key/3))) for key in keys]
```

```python
x = np.arange(len(labels))  # label positions
width = 0.4                 # bar width

plt.figure()

plt.bar(x - width/2, train, width, label="Train")
plt.bar(x + width/2, test,  width, label="Test")

plt.xticks(x, labels)
plt.title("Train vs Test Comparison")
plt.legend()

plt.tight_layout()
plt.show()
```
